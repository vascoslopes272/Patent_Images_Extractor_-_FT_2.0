# 02 — Phase 2: Resize + Pad for DINOv2
Takes crops from Phase 1, resizes with aspect ratio preserved, pads to square.
Output goes to processed/ — these are DINOv2-ready images.

```python
import sys; sys.path.insert(0, "..")
import yaml, json
from pathlib import Path
from src.processor import process_crops

with open("../config.yaml") as f:
    cfg = yaml.safe_load(f)

# Load crop index from Phase 1
index_path = Path(cfg["paths"]["logs"]) / "crop_index.json"
with open(index_path) as f:
    crop_index = json.load(f)

all_crop_paths = [p for paths in crop_index.values() for p in paths]
print(f"Total crops to process: {len(all_crop_paths)}")
print(f"Target size: {cfg["processor"]["target_size"]}x{cfg["processor"]["target_size"]} px")
```

```python
# Run processing — skips already-processed files
processed_paths = process_crops(all_crop_paths, cfg)
```

```python
# Quick visual check — show first 6 processed images
import matplotlib.pyplot as plt
from PIL import Image

fig, axes = plt.subplots(2, 3, figsize=(12, 8))
for ax, p in zip(axes.flat, processed_paths[:6]):
    img = Image.open(p)
    ax.imshow(img)
    ax.set_title(Path(p).name[:30], fontsize=7)
    ax.axis("off")
plt.tight_layout()
plt.show()
```
