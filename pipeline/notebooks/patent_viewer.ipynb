# Patent Viewer & Selector

Browse the PatSeer dataset, inspect extracted crops, and flag patents for inclusion in the pipeline.

**Independent of pipeline phases** — run at any time, before or after extraction.

- Edit the `filters` dict in the cell below to narrow what you see (no need to touch `config.yaml`)
- Use **← →** to navigate, **Select / Deselect** to flag patents
- Every click saves immediately to `logs/selected_patents.json`
- Run `ui.export()` to also write `logs/selected_patents.xlsx`

**To feed your selection into the pipeline:** set `subset.mode: "selected"` in `config.yaml`.

```python
import sys; sys.path.insert(0, "..")
from src.config_loader import load_config
from src.selector import PatentSelectorUI

cfg = load_config()

# ── What to show (edit here — no need to touch config.yaml) ──────────────
filters = {
    "cpc_first":    ["B64C29/0041", "B64C29/0083", "B64C29/016"],  # None = show all
    "legal_status": None,   # "ALIVE" | "DEAD" | None
    "record_type":  None,   # "Patent" | None
}
# Set filters=None to browse the entire dataset without any filtering.

ui = PatentSelectorUI(cfg, filters=filters)
ui.show()
```

```python
# How many patents are currently selected?
ui.summary()
```

```python
# Persist JSON + write logs/selected_patents.xlsx for inspection
ui.export()
```
